```python
#rename Files
import os

# Path to your folder
folder_path = "/content/drive/MyDrive/Dataset/Naming/Down/Down_non-occluded"  # Change to your folder path

# Get all files in the folder (excluding directories)
files = [f for f in os.listdir(folder_path) if os.path.isfile(os.path.join(folder_path, f))]

# Sort to ensure consistent order
files.sort()

# Rename files
for i, filename in enumerate(files):
    ext = os.path.splitext(filename)[1]  # Keep original extension
    new_name = f"Down_non_occluded_0000{i+1}{ext}"

    old_path = os.path.join(folder_path, filename)
    new_path = os.path.join(folder_path, new_name)

    os.rename(old_path, new_path)

print("All files renamed.")
```

```python
#merging two folders
import os
import shutil

# Define source folders and destination folder
src_folder_1 = "/content/drive/MyDrive/Dataset/Naming/Down/Down_non-occluded"  # change to your actual folder path
src_folder_2 = "/content/drive/MyDrive/Dataset/Naming/Down/Down_occluded"
dest_folder = "/content/drive/MyDrive/Dataset/Final DataSet/Down"

# Create destination folder if it doesn't exist
os.makedirs(dest_folder, exist_ok=True)

# Function to copy files from source to destination
def copy_files(src, dst):
    for filename in os.listdir(src):
        src_path = os.path.join(src, filename)
        dst_path = os.path.join(dst, filename)

        # If file with same name exists, rename to avoid overwrite
        if os.path.exists(dst_path):
            base, ext = os.path.splitext(filename)
            counter = 1
            while os.path.exists(dst_path):
                new_filename = f"{base}_{counter}{ext}"
                dst_path = os.path.join(dst, new_filename)
                counter += 1

        shutil.copy2(src_path, dst_path)

# Copy files from both folders
copy_files(src_folder_1, dest_folder)
copy_files(src_folder_2, dest_folder)

print("Files merged into:", dest_folder)
```
